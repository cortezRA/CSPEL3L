```python
raw_data = pd.read_csv("Coca-Cola_stock_history.csv")
```

```python
df_fullstock = raw_data.copy()
```

```python
import matplotlib.pyplot as plt
```

```python
df_fullstock.Open.plot(figsize=(20,5) ,title="Opening Stock Prices of Cola-Cola from 1962-2022")
plt.show()
```

```python
df_fullstock.High.plot(figsize=(20,5) ,title="Highest Stock Prices of Cola-Cola from 1962-2022")
plt.show()
```

```python
df_fullstock.High.plot(figsize=(20,5) ,title="Highest Stock Prices of Cola-Cola from 1962-2022")
df_fullstock.Low.plot(figsize=(20,5) ,title="Lowest Stock Prices of Cola-Cola from 1962-2022")
plt.title("Highest vs Lowest Stock Prices")
plt.show()
```
